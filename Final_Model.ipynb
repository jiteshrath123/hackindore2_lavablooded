```python
import numpy as np
import pandas as pd
```

```python
my_df = pd.read_csv('merge_pivot.csv')
```

```python
my_df.set_index('movieId',inplace = True)
```

```python
my_df.head()
```

```python
# my_input = my_df.iloc[12633]
my_input = my_df.iloc[1]
```

```python
my_input
```

```python
computation = (((my_df - my_input)**2).sum(axis = 1)).sort_values(ascending = True)
```

```python
computation.head()
```

computation.head()
# new_com = computation[:]

```python
count = 0
my_df1 = pd.read_csv('movies.csv')
my_df1.set_index('movieId',inplace=True)
for val in computation.index:
    if count > 20:
        break
    try:
        print(my_df1.iloc[val]['title'])
        count += 1
    except:
        pass
```
